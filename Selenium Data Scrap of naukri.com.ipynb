```python
import selenium
import pandas as pd
from selenium import webdriver
import time
# Opening Browsers in Incognito using Options
from selenium.webdriver.chrome.options import Options
chrome_options = Options()
chrome_options.add_argument('--incognito')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException,ElementNotInteractableException,InvalidArgumentException,InvalidCookieDomainException
```

```python
def naukri(url):
    driver = webdriver.Chrome(r"C:\Users\hp\OneDrive\Desktop\chromedriver.exe")
    # 1. first get the webpage https://www.naukri.com/
    driver.get(url)
    
    # 2. Enter “Data Analyst” in “Skill,Designations,Companies” field 
    job_search=driver.find_element_by_xpath('//input[@class="suggestor-input "]').send_keys("Data Analyst")
    
    # and enter “Bangalore” in “enter the location” field 
    location_search=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[3]/div/div/div/input').send_keys("Banglore")
 
    # 3. Then click the search button
    search_buton=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[6]').click()
    time.sleep(5)

    # 4. Then scrape the data for the first 10 jobs results you get.
    job_titles=[]
    title_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")[:10]
    for i in title_tags:
        job_titles.append(i.text)
    
    company_names=[]
    company_tags=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")[:10]
    for i in company_tags:
        company_names.append(i.text)
        
    experience_list=[]
    exp_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']/span")[:10]
    for i in exp_tags:
        experience_list.append(i.text)
        
    salary_list=[]
    sal_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi salary']/span")[:10]
    for i in sal_tags:
        salary_list.append(i.text)
        
    locations_list=[]
    loc_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span")[:10]
    for i in loc_tags:
        locations_list.append(i.text)
        
    jobs=pd.DataFrame({})
    jobs['job_titles']=job_titles
    jobs['company_names']=company_names
    jobs['locations_list']=locations_list
    jobs['experience_list']=experience_list
    jobs['salary_list']=salary_list
    
    driver.close()
    return(jobs)
```
